```python
import pandas as pd
```

```python
# Define file paths and number of top companies
input_file = 'results/intermediate_results.full.json'

companies_data = pd.read_json(input_file)
companies_data = companies_data.dropna(subset=['ideaValidatorScore', 'ideaValidatorSummary'])
top_companies = companies_data.sort_values(by='ideaValidatorScore', ascending=False).head(10)
```

```python
top_companies
```

```python
pd.set_option('max_colwidth', None)
top_10_comp = top_companies[['company_name', 'short_description', 'long_description', 'ideaValidatorSummary', 'ideaValidatorScore']]
```

```python
top_10_comp.to_csv('yc-wc24-top-10-companies.csv', sep='\t')
```
